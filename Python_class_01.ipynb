```python
def LargestPair(num):
    num_str = str(num)
    max_pair = 0

    for i in range(len(num_str) - 1):
        pair = int(num_str[i:i+2])  # 2-digit pair

        if pair > max_pair:
            max_pair = pair

    return max_pair


# Example
print(LargestPair(4759472))  # Output: 94
```

```python
def Primes(num):
    if num <= 1:
        return "false"

    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return "false"

    return "true"


# Example
print(Primes(7))   # true
print(Primes(10))  # false
```
